```python
#property_images.xlsx demo_myday

import pandas as pd
from datetime import datetime


final_jpg_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_jpg_file.xlsx'
final_combined_properties_data_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx'


final_jpg_file = pd.read_excel(final_jpg_file_path)
final_combined_properties_data = pd.read_excel(final_combined_properties_data_path)


property_images = pd.DataFrame()


property_images['id'] = range(1000, len(final_jpg_file) + 1000)


property_images['property_id'] = final_jpg_file['Property Name'].map(
    final_combined_properties_data.set_index('Reference')['properties_row']
)


property_images['property_image'] = final_jpg_file['modified link']


property_images['image_name'] = final_jpg_file['image name']


property_images['file_size'] = final_jpg_file['file size']


current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


property_images['created_at'] = current_timestamp
property_images['updated_at'] = current_timestamp


output_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\property_images.xlsx'
property_images.to_excel(output_path, index=False)

print(f"Processed data has been saved to {output_path}")
```
