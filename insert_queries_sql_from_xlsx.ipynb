```python
#single insert_queries sql file for allexcel files  

import pandas as pd
import os

# Define the folder path where your Excel files are stored
folder_path = r"C:\Users\MSI\Documents\Migration Tool New\03 Demo 01 Myday DB\02 demo_myday_db_v1\outputs"
output_sql_file = r"C:\Users\MSI\Documents\Migration Tool New\03 Demo 01 Myday DB\02 demo_myday_db_v1\outputs\insert_queries.sql"

# Function to generate insert query for each row of the DataFrame
def generate_insert_query(table_name, df):
    columns = ', '.join([f'`{col}`' for col in df.columns])
    values_list = []

    for _, row in df.iterrows():
        values = ', '.join([f"'{str(val)}'" if val is not None else 'NULL' for val in row])
        values_list.append(f"INSERT INTO {table_name} ({columns}) VALUES ({values});")

    return "\n".join(values_list)

# Open a file to save all the insert queries
with open(output_sql_file, 'w', encoding='utf-8') as sql_file:
    # Loop through all the Excel files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.xlsx'):
            file_path = os.path.join(folder_path, file_name)
            # Read the Excel file
            df = pd.read_excel(file_path)
            
            # Use the file name (without extension) as the table name
            table_name = os.path.splitext(file_name)[0]
            
            # Generate the insert queries
            insert_queries = generate_insert_query(table_name, df)
            
            # Write the queries to the .sql file
            sql_file.write(f"-- Insert queries for {table_name}\n")
            sql_file.write(insert_queries)
            sql_file.write("\n\n")

print(f"Insert queries saved to {output_sql_file}")
```

```python
#it generates a separate .sql file for each .xlsx file and saves them in a new folder named insert_sql inside the current folder,



import os
import pandas as pd

# Define the folder path where your Excel files are stored
folder_path = r"C:\Users\MSI\Documents\Migration Tool New\03 Demo 01 Myday DB\02 demo_myday_db_v1\outputs"

# Create a new folder named 'insert_sql' inside the current folder for storing the generated .sql files
output_folder = os.path.join(folder_path, 'insert_sql')
os.makedirs(output_folder, exist_ok=True)

# Function to generate insert query for each row of the DataFrame
def generate_insert_query(table_name, df):
    columns = ', '.join([f'`{col}`' for col in df.columns])
    values_list = []

    for _, row in df.iterrows():
        values = ', '.join([f"'{str(val)}'" if val is not None else 'NULL' for val in row])
        values_list.append(f"INSERT INTO {table_name} ({columns}) VALUES ({values});")

    return "\n".join(values_list)

# Loop through all the Excel files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file_name)
        # Read the Excel file
        df = pd.read_excel(file_path)
        
        # Use the file name (without extension) as the table name
        table_name = os.path.splitext(file_name)[0]
        
        # Generate the insert queries
        insert_queries = generate_insert_query(table_name, df)
        
        # Create an individual .sql file for each Excel file in the 'insert_sql' folder
        output_sql_file = os.path.join(output_folder, f"{table_name}.sql")
        with open(output_sql_file, 'w', encoding='utf-8') as sql_file:
            # Write the queries to the .sql file
            sql_file.write(f"-- Insert queries for {table_name}\n")
            sql_file.write(insert_queries)
        
        print(f"Insert queries saved to {output_sql_file}")

print(f"All insert queries have been saved to the 'insert_sql' folder.")
```
