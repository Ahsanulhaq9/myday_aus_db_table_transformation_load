```python
#contact_communications  v2

import pandas as pd
from datetime import datetime


file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
df = pd.read_excel(file_path)


communication_data = df['Inner_Communication'].apply(lambda x: x if pd.notnull(x) else 'None')


current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


data = {
    #'id': range(1, len(communication_data) + 1),
    'id': range(1000001, 1000001 + len(communication_data)),  # Auto-increment starting from 1000001
    'contact_id': range(1, len(communication_data) + 1),
    'communication': communication_data,
    'created_at': current_time,  
    'updated_at': current_time,  
    'contact_details_id': None  
}


new_df = pd.DataFrame(data)


output_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\contact_communications.xlsx'
new_df.to_excel(output_file_path, index=False)

print(f"File saved successfully to {output_file_path}")
```
