```python
#v2 demo_myday

import pandas as pd
from datetime import datetime


contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
properties_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx'

contacts_df = pd.read_excel(contacts_file)
properties_df = pd.read_excel(properties_file)


tenant_df = contacts_df[
    (contacts_df['Roles'] == 'Tenant') |
    (contacts_df['Roles'] == 'Supplier, Tenant') |
    (contacts_df['Roles'] == 'Owner, Supplier, Tenant') |
    (contacts_df['Roles'] == 'Owner, Tenant')
].copy()


tenant_contacts = pd.DataFrame()


tenant_contacts['id'] = range(1000001, 1000001 + len(tenant_df))


tenant_contacts['reference'] = tenant_df['Reference'].values


tenant_contacts['contact_id'] = tenant_df['Row No'].values


tenant_contacts['property_id'] = tenant_contacts['reference'].apply(
    lambda ref: properties_df[properties_df['Tenant'] == ref]['properties_row'].values[0]
    if ref in properties_df['Tenant'].values else None
)


tenant_contacts[['first_name', 'last_name']] = tenant_contacts['reference'].str.split(n=1, expand=True)


tenant_contacts['salutation'] = tenant_df['Inner_Salutation'].values


tenant_contacts['company_name'] = None


tenant_contacts['mobile_phone'] = tenant_df['Inner_Mobile'].values


tenant_contacts['work_phone'] = tenant_df['Inner_Work'].values


tenant_contacts['home_phone'] = tenant_df['Inner_Mobile 2'].values


tenant_contacts['email'] = tenant_df['Inner_Email'].values


tenant_contacts['abn'] = tenant_df['Inner_ABN'].values


tenant_contacts['notes'] = None


tenant_contacts['company_id'] = 1


tenant_contacts['user_id'] = None


current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
tenant_contacts['created_at'] = current_time
tenant_contacts['updated_at'] = current_time


tenant_contacts['status'] = True


output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\tenant_contacts.xlsx'
tenant_contacts.to_excel(output_file, index=False)

print(f"File saved as {output_file}")
```
