```python
#v2 demo_myday

import pandas as pd
from datetime import datetime


contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
sales_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\properties_sales_data.xlsx'
seller_contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\seller_contacts.xlsx'

contacts_data = pd.read_excel(contacts_file)
sales_data = pd.read_excel(sales_file)
seller_contacts_data = pd.read_excel(seller_contacts_file)


seller_properties = pd.DataFrame(columns=['id', 'seller_contact_id', 'property_id', 'created_at', 'updated_at'])


seller_properties['id'] = range(1000, len(sales_data) + 1000)


reference_to_id = seller_contacts_data.set_index('reference')['id'].to_dict()


seller_properties['seller_contact_id'] = sales_data['Seller'].map(reference_to_id)


seller_properties['property_id'] = sales_data['Row']


current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
seller_properties['created_at'] = current_time
seller_properties['updated_at'] = current_time


output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\seller_properties.xlsx'
seller_properties.to_excel(output_file, index=False)

print("seller_properties.xlsx has been created successfully.")
```
