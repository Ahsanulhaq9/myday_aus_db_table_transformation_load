```python
#optional_properties

import pandas as pd
from datetime import datetime


properties_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\properties.xlsx'
properties_df = pd.read_excel(properties_file_path)


optional_properties_df = pd.DataFrame()


optional_properties_df['id'] = range(1, len(properties_df) + 1)


optional_properties_df['property_id'] = properties_df['id']


optional_properties_df['garages'] = properties_df['car_space']


optional_properties_df['carports'] = None
optional_properties_df['open_car_space'] = None


current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


optional_properties_df['created_at'] = current_timestamp
optional_properties_df['updated_at'] = current_timestamp


output_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\optional_properties.xlsx'
optional_properties_df.to_excel(output_file_path, index=False)

print(f'optional_properties.xlsx created successfully at {output_file_path}')
```
