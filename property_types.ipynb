```python
#property_types

import pandas as pd
from datetime import datetime


file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx'
df = pd.read_excel(file_path)


unique_property_types = df['Property Type '].unique()


data = {
    'id': range(1, len(unique_property_types) + 1),
    'type': unique_property_types,
    'created_at': [datetime.now().strftime('%Y-%m-%d %H:%M:%S')] * len(unique_property_types),
    'updated_at': [datetime.now().strftime('%Y-%m-%d %H:%M:%S')] * len(unique_property_types)
}


property_types_df = pd.DataFrame(data)


output_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\property_types.xlsx'


property_types_df.to_excel(output_file_path, index=False)

print(f"File created successfully at {output_file_path}")
```
