```python
#owner_contacts final


import pandas as pd
from datetime import datetime


contacts_df = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx')
properties_df = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx')
owner_folios_df = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\owner_folios.xlsx')


owner_contact_df = pd.DataFrame(columns=[
    'id', 'contact_id', 'property_id', 'reference', 'first_name', 'last_name',
    'salutation', 'company_name', 'mobile_phone', 'work_phone', 'home_phone', 
    'email', 'notes', 'company_id', 'user_id', 'created_at', 'updated_at', 
    'status', 'owner_folio_id'
])


id_counter = 1000001


current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


for idx, row in properties_df[properties_df['Owner'].notna()].iterrows():
    owner_data = row['Owner'].split()
    first_name = owner_data[0] if len(owner_data) > 0 else ''
    last_name = ' '.join(owner_data[1:]) if len(owner_data) > 1 else ''
    
    
    contact_match = contacts_df[contacts_df['Reference'] == row['Owner']]
    
    if not contact_match.empty:
        contact_id = contact_match.iloc[0]['Row No']
        salutation = contact_match.iloc[0]['Inner_Salutation']
        mobile_phone = contact_match.iloc[0]['Phone']
        work_phone = contact_match.iloc[0]['Inner_Work']
        email = contact_match.iloc[0]['Inner_Email']
    else:
        contact_id = None
        salutation = None
        mobile_phone = None
        work_phone = None
        email = None
    
    
    if contact_id is None:
        continue
    
    
    new_row = {
        'id': id_counter,
        'contact_id': contact_id,
        'property_id': row['properties_row'],
        'reference': row['Owner'],
        'first_name': first_name,
        'last_name': last_name,
        'salutation': salutation,
        'company_name': None,
        'mobile_phone': mobile_phone,
        'work_phone': work_phone,
        'home_phone': None,
        'email': email,
        'notes': None,
        'company_id': 1,
        'user_id': None,
        'created_at': current_time,
        'updated_at': current_time,
        'status': 1,
        'owner_folio_id': None  
    }
    
    
    folio_match = owner_folios_df[(owner_folios_df['owner_contact_id'].notna()) & 
                                  (owner_folios_df['owner_contact_id'] == id_counter)]
    if not folio_match.empty:
        new_row['owner_folio_id'] = folio_match.iloc[0]['Id']
    else:
        new_row['owner_folio_id'] = 'NULL'

    
    owner_contact_df = pd.concat([owner_contact_df, pd.DataFrame([new_row])], ignore_index=True)

    
    id_counter += 1


owner_contact_df.to_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\owner_contacts.xlsx', index=False)

print("owner_contacts.xlsx has been created successfully!")
```
