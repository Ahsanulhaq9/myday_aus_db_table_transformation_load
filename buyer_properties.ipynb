```python
#final demo_myday

import pandas as pd
from datetime import datetime


buyer_contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\buyer_contacts.xlsx'


buyer_contacts = pd.read_excel(buyer_contacts_file)


filtered_buyer_contacts = buyer_contacts[buyer_contacts['property_id'].notnull() & (buyer_contacts['property_id'] != 0)]

buyer_properties_data = {
    'id': range(1, len(filtered_buyer_contacts) + 1),
    'buyer_contact_id': filtered_buyer_contacts['id'],  # Taking 'id' from buyer_contacts
    'property_id': filtered_buyer_contacts['property_id'],
    'created_at': [datetime.now().strftime('%Y-%m-%d %H:%M:%S')] * len(filtered_buyer_contacts),
    'updated_at': [datetime.now().strftime('%Y-%m-%d %H:%M:%S')] * len(filtered_buyer_contacts)
}


buyer_properties_df = pd.DataFrame(buyer_properties_data)


output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\buyer_properties.xlsx'
buyer_properties_df.to_excel(output_file, index=False)

print(f'File saved as {output_file}')
```
