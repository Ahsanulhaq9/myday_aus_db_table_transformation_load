```python
#v2

import pandas as pd
from datetime import datetime


input_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
df = pd.read_excel(input_file)


supplier_df = df[
    (df['Roles'] == 'Supplier') |
    (df['Roles'] == 'Supplier, Tenant') |
    (df['Roles'] == 'Owner, Supplier') |
    (df['Roles'] == 'Owner, Supplier, Tenant') |
    (df['Roles'] == 'Seller, Supplier')
].copy()


supplier_contacts = pd.DataFrame()


supplier_contacts['id'] = range(1000001, 1000001 + len(supplier_df))


supplier_contacts['contact_id'] = supplier_df['Row No'].values


supplier_contacts['reference'] = supplier_df['Reference'].values


supplier_contacts[['first_name', 'last_name']] = supplier_contacts['reference'].str.split(n=1, expand=True)


supplier_contacts['salutation'] = supplier_df['Inner_Salutation'].values


supplier_contacts['company_name'] = None


supplier_contacts['mobile_phone'] = supplier_df['Inner_Mobile'].values


supplier_contacts['work_phone'] = supplier_df['Inner_Work'].values


supplier_contacts['home_phone'] = supplier_df['Inner_Mobile 2'].values


supplier_contacts['email'] = supplier_df['Inner_Email'].values


supplier_contacts['notes'] = None


supplier_contacts['company_id'] = 1


supplier_contacts['user_id'] = None


supplier_contacts['abn'] = supplier_df['Inner_ABN'].values


current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
supplier_contacts['created_at'] = current_time
supplier_contacts['updated_at'] = current_time


output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\supplier_contacts.xlsx'
supplier_contacts.to_excel(output_file, index=False)

print(f"File saved as {output_file}")
```
