```python
#v4 final demo_myday


import pandas as pd
from datetime import datetime


properties_sales_data = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\properties_sales_data.xlsx')
seller_contacts = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\seller_contacts.xlsx')
buyer_contacts = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\buyer_contacts.xlsx')


rows = []


current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


start_id = 100


for index, row in properties_sales_data.iterrows():
    
    property_id = row['Row']
    if pd.isna(property_id):
        continue
    
    
    seller_ref = row['Seller']
    seller_id = seller_contacts[seller_contacts['reference'] == seller_ref]['id'].values
    seller_id = seller_id[0] if len(seller_id) > 0 else 0

   
    buyer_ref = row['Buyer']
    buyer_id = buyer_contacts[buyer_contacts['reference'] == buyer_ref]['id'].values
    buyer_id = buyer_id[0] if len(buyer_id) > 0 else 0

    
    has_buyer = 'true' if pd.notna(row['Buyer']) else 'false'

    
    status = 1

    
    rows.append({
        'id': start_id + index,  # Auto-incremental ID starting from 1
        'property_id': property_id,
        'seller_id': seller_id,
        'buyer_id': buyer_id,
        'has_buyer': has_buyer,
        'created_at': current_timestamp,
        'updated_at': current_timestamp,
        'status': status
    })


property_sales_agreements_df = pd.DataFrame(rows)


output_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\property_sales_agreements.xlsx'
property_sales_agreements_df.to_excel(output_path, index=False)

print(f"File created successfully at {output_path}")
```
