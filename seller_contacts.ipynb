```python
#seller_contacts


import pandas as pd
from datetime import datetime


contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
sales_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\properties_sales_data.xlsx'


contacts_df = pd.read_excel(contacts_file)
sales_df = pd.read_excel(sales_file)


seller_df = contacts_df[
    (contacts_df['Roles'] == 'Seller') | 
    (contacts_df['Roles'] == 'Owner, Seller') | 
    (contacts_df['Roles'] == 'Seller, Supplier')
].copy()


seller_contacts = pd.DataFrame()


#seller_contacts['id'] = range(1, len(seller_df) + 1)


seller_contacts['id'] = range(1000001, 1000001 + len(seller_df))



seller_contacts['contact_id'] = seller_df['Row No'].values


property_ids = []
for reference in seller_df['Reference']:
    matching_row = sales_df[sales_df['Seller'] == reference]
    if not matching_row.empty:
        property_ids.append(matching_row['Row'].values[0])
    else:
        property_ids.append(None)
seller_contacts['property_id'] = property_ids


seller_contacts['reference'] = seller_df['Reference'].values


seller_contacts[['first_name', 'last_name']] = seller_contacts['reference'].str.split(n=1, expand=True)


seller_contacts['last_name'] = seller_contacts['last_name'].fillna('')


seller_contacts['salutation'] = seller_df['Inner_Salutation'].values


seller_contacts['company_name'] = None


seller_contacts['mobile_phone'] = seller_df['Inner_Mobile'].values


seller_contacts['work_phone'] = seller_df['Inner_Work'].values


seller_contacts['home_phone'] = seller_df['Inner_Mobile 2'].values


seller_contacts['email'] = seller_df['Inner_Email'].values


seller_contacts['notes'] = None


seller_contacts['abn'] = seller_df['Inner_ABN'].values


seller_contacts['company_id'] = 1


seller_contacts['user_id'] = 'NULL'


current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
seller_contacts['created_at'] = current_time
seller_contacts['updated_at'] = current_time


output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\seller_contacts.xlsx'
seller_contacts.to_excel(output_file, index=False)

print(f"File saved as {output_file}")
```
