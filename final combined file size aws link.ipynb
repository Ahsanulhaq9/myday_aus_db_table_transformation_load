```python
#jpg 3 files combine to final jpg


import pandas as pd

# File paths
source_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\Source Data\Aws image pdf source\extracted_files_info_jpg.xlsx'
jpg_details_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\Source Data\Aws image pdf source\jpg_file_details.xlsx'
s3_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\Source Data\Aws image pdf source\s3_uploaded_files_jpg.xlsx'
output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\Source Data\Aws image pdf source\final_jpg_file.xlsx'

# Load all Excel files into DataFrames
source_df = pd.read_excel(source_file)
jpg_details_df = pd.read_excel(jpg_details_file)
s3_df = pd.read_excel(s3_file)

# Step 1: Create 'image name' column by removing 'E:\Extracted_Files\' from the 'File Path' column
source_df['image name'] = source_df['File Path'].str.replace(r'E:\\Extracted_Files\\', '', regex=True)

# Step 2: Merge with jpg_file_details.xlsx to add 'file size'
merged_df = pd.merge(source_df, jpg_details_df[['File Name', 'File Size (bytes)']], 
                     left_on='image name', right_on='File Name', how='left')

# Step 3: Rename 'File Size (bytes)' to 'file size'
merged_df.rename(columns={'File Size (bytes)': 'file size'}, inplace=True)

# Step 4: Compare 'image name' with 'File Name' in s3_uploaded_files_jpg.xlsx and add 'aws image link'
merged_df = pd.merge(merged_df, s3_df[['File Name', 'Object URL']], 
                     left_on='image name', right_on='File Name', how='left')

# Step 5: Rename 'Object URL' to 'aws image link'
merged_df.rename(columns={'Object URL': 'aws image link'}, inplace=True)

# Step 6: Create 'modified link' column by removing 'https://mydaybucketlive.s3.ap-southeast-2.amazonaws.com/' from 'aws image link'
merged_df['modified link'] = merged_df['aws image link'].str.replace(
    r'https://mydaybucketlive.s3.ap-southeast-2.amazonaws.com/', '', regex=True)

# Save the final DataFrame to a new Excel file
merged_df.to_excel(output_file, index=False)

print(f"Final data has been saved to {output_file}")
```
