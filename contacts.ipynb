```python
#contacts table

import pandas as pd
from datetime import datetime


file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
df = pd.read_excel(file_path)


contacts_df = pd.DataFrame()

# Create an 'id' column with auto-incremental values
#contacts_df['id'] = range(1, len(df) + 1)
contacts_df['id'] = range(1000001, 1000001 + len(df))


contacts_df['reference'] = df['Reference']


contacts_df['type'] = None


contacts_df[['first_name', 'last_name']] = contacts_df['reference'].str.split(n=1, expand=True)


contacts_df['salutation'] = df['Inner_Salutation']
contacts_df['company_name'] = df['Inner_Company']
contacts_df['mobile_phone'] = df['Inner_Mobile']
contacts_df['work_phone'] = df['Inner_Work']
contacts_df['home_phone'] = None  # Set home_phone as None
contacts_df['email'] = df['Inner_Email']
contacts_df['abn'] = df['Inner_ABN']
contacts_df['notes'] = None  # Set notes as None


def role_column(role, role_column):
    return role_column.apply(lambda x: 1 if role in str(x).lower() else 0)

contacts_df['owner'] = role_column('owner', df['Roles'])
contacts_df['tenant'] = role_column('tenant', df['Roles'])
contacts_df['supplier'] = role_column('supplier', df['Roles'])
contacts_df['seller'] = role_column('seller', df['Roles'])


contacts_df['company_id'] = 1


current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
contacts_df['created_at'] = current_timestamp
contacts_df['updated_at'] = current_timestamp


contacts_df['buyer'] = df['Inner_Buyer'].apply(lambda x: 0 if pd.isna(x) or str(x).strip() == '' else 1)


output_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\contacts.xlsx'


contacts_df.to_excel(output_file_path, index=False)


print(f"The contacts.xlsx file has been created successfully and saved to {output_file_path}")
```
