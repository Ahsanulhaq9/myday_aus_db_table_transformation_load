```python
#buyer_contacts.xlsx  final

import pandas as pd
from datetime import datetime


contacts_df = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx')
properties_df = pd.read_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx')


buyer_df = contacts_df[contacts_df['Inner_Buyer'].notna()].copy()


def match_value(value, column):
   
    full_match = column[column == value]
    if not full_match.empty:
        return full_match.index[0]

    
    partial_match = column[column.str.contains(value, na=False, regex=False)]
    if not partial_match.empty:
        return partial_match.index[0]
    return None


data = []


auto_id = 1


for index, row in buyer_df.iterrows():
    contact_id = None
    property_id = None
    
    
    matched_contact_index = match_value(row['Inner_Buyer'], contacts_df['Reference'])
    if matched_contact_index is not None:
        contact_id = contacts_df.loc[matched_contact_index, 'Row No']

    
    if pd.notna(row['Inner_Property (Buyer)']):
        matched_property_index = match_value(row['Inner_Property (Buyer)'], properties_df['Reference'])
        if matched_property_index is not None:
            property_id = properties_df.loc[matched_property_index, 'properties_row']

    
    if property_id is None:
        continue

    
    reference_value = row['Inner_Buyer']
    first_name = reference_value.split()[0] if len(reference_value.split()) > 0 else None
    last_name = reference_value.split()[1] if len(reference_value.split()) > 1 else None

    
    salutation = contacts_df.loc[matched_contact_index, 'Inner_Salutation'] if matched_contact_index is not None else None
    company_name = contacts_df.loc[matched_contact_index, 'Inner_Company'] if matched_contact_index is not None else None
    mobile_phone = contacts_df.loc[matched_contact_index, 'Inner_Mobile'] if matched_contact_index is not None else None
    work_phone = contacts_df.loc[matched_contact_index, 'Inner_Work'] if matched_contact_index is not None else None
    home_phone = contacts_df.loc[matched_contact_index, 'Inner_Mobile 2'] if matched_contact_index is not None else None
    email = contacts_df.loc[matched_contact_index, 'Inner_Email'] if matched_contact_index is not None else None
    abn = contacts_df.loc[matched_contact_index, 'Inner_ABN'] if matched_contact_index is not None else None

    
    data.append([
        auto_id,  # id (auto-incremental)
        contact_id,  # contact_id
        property_id,  # property_id
        reference_value,  # reference
        first_name,  # first_name
        last_name,  # last_name
        salutation,  # salutation
        company_name,  # company_name
        mobile_phone,  # mobile_phone
        work_phone,  # work_phone
        home_phone,  # home_phone
        email,  # email
        None,  # notes (NULL for all)
        1,  # company_id (1 for all)
        'NULL',  # user_id (NULL for all)
        datetime.now().strftime('%Y-%m-%d %H:%M:%S'),  
        datetime.now().strftime('%Y-%m-%d %H:%M:%S'),  
        abn  # abn
    ])

    
    auto_id += 1


columns = [
    'id', 'contact_id', 'property_id', 'reference', 'first_name', 'last_name',
    'salutation', 'company_name', 'mobile_phone', 'work_phone', 'home_phone', 
    'email', 'notes', 'company_id', 'user_id', 'created_at', 'updated_at', 'abn'
]


buyer_contacts_df = pd.DataFrame(data, columns=columns)
buyer_contacts_df.to_excel(r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\buyer_contacts.xlsx', index=False)

print("buyer_contacts.xlsx file has been created successfully.")
```
