```python
#v2 demo_myday


import pandas as pd
from datetime import datetime
import os


contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
tenant_contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\tenant_contacts.xlsx'
properties_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx'


if os.path.exists(contacts_file) and os.path.exists(tenant_contacts_file) and os.path.exists(properties_file):
    
    contacts_data = pd.read_excel(contacts_file)
    tenant_contacts_data = pd.read_excel(tenant_contacts_file)
    properties_data = pd.read_excel(properties_file)

    
    tenant_properties = pd.DataFrame(columns=['id', 'tenant_contact_id', 'property_id', 'created_at', 'updated_at', 'status'])

    
    tenant_properties['id'] = range(10000, len(properties_data) + 10000)

    
    merged_data = pd.merge(
        properties_data[['Tenant', 'properties_row']],  
        tenant_contacts_data[['reference', 'id']], 
        left_on='Tenant',  
        right_on='reference',
        how='left'  
    )

    
    tenant_properties['tenant_contact_id'] = merged_data['id']

    
    tenant_properties['property_id'] = merged_data['properties_row']

    
    current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    tenant_properties['created_at'] = current_time
    tenant_properties['updated_at'] = current_time

    
    tenant_properties['status'] = 1

    
    tenant_properties = tenant_properties.dropna(subset=['tenant_contact_id'])

    
    output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\tenant_properties.xlsx'
    tenant_properties.to_excel(output_file, index=False)

    print("tenant_properties.xlsx has been created successfully.")
else:
    print("One or more of the input files do not exist.")
```
