```python
#v3 demo_myday

import pandas as pd
from datetime import datetime
import os


contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
owner_contacts_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\owner_contacts.xlsx'
properties_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx'


if os.path.exists(contacts_file) and os.path.exists(owner_contacts_file) and os.path.exists(properties_file):
    
    contacts_data = pd.read_excel(contacts_file)
    owner_contacts_data = pd.read_excel(owner_contacts_file)
    properties_data = pd.read_excel(properties_file)

    
    owner_properties = pd.DataFrame(columns=['id', 'owner_contact_id', 'property_id', 'created_at', 'updated_at'])

    
    owner_properties['id'] = range(10000, len(properties_data) + 10000)

    
    merged_data = pd.merge(
        properties_data[['Owner', 'properties_row']],  
        owner_contacts_data[['reference', 'id']],  
        left_on='Owner',  
        right_on='reference',
        how='left'  
    )

    
    owner_properties['owner_contact_id'] = merged_data['id']

    
    owner_properties['property_id'] = merged_data['properties_row']

    
    current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    owner_properties['created_at'] = current_time
    owner_properties['updated_at'] = current_time

    
    owner_properties = owner_properties.dropna(subset=['owner_contact_id'])

    
    output_file = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\owner_properties.xlsx'
    owner_properties.to_excel(output_file, index=False)

    print("owner_properties.xlsx has been created successfully.")
else:
    print("One or more of the input files do not exist.")
```
