```python
#countries final

import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime


url = "https://en.wikipedia.org/wiki/List_of_sovereign_states"


response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')


table = soup.find('table', {'class': 'sortable'})


countries = []


for row in table.find_all('tr')[1:]:  # Skip the header row
    cells = row.find_all('td')
    if len(cells) > 1:
        country_name = cells[0].find('a').text.strip()  # The country name is in the first <a> tag
        countries.append(country_name)

countries = sorted(countries)


current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


df = pd.DataFrame({
    'id': range(1, len(countries) + 1),
    'country_name': countries,
    'created_at': current_timestamp,
    'updated_at': current_timestamp
})


output_file = 'countries.xlsx'
df.to_excel(output_file, index=False)

print(f"Alphabetical country names with timestamps have been saved to {output_file}")
```
