```python
#contact_labels

import pandas as pd
from datetime import datetime

# Load the Excel file
file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_contacts_data.xlsx'
df = pd.read_excel(file_path)

# Extract the 'Labels' column
labels_data = df['Labels']

# Generate current timestamp
current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

# Prepare the data for the new DataFrame
data = {
    'id': range(1, len(labels_data) + 1),
    'contact_id': range(1, len(labels_data) + 1),
    'labels': labels_data,
    'created_at': current_time,  # Assigning the current timestamp
    'updated_at': current_time   # Assigning the current timestamp
}

# Create a new DataFrame
new_df = pd.DataFrame(data)

# Save the DataFrame to a new Excel file
output_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\contact_labels.xlsx'
new_df.to_excel(output_file_path, index=False)

print(f"File saved successfully to {output_file_path}")
```
