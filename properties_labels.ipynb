```python
#peoperty_labels with timestamp


import pandas as pd
import re
from datetime import datetime


file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\final_combined_properties_data.xlsx'
df = pd.read_excel(file_path)


property_labels = df[['Property Labels', 'properties_row']].dropna()


split_labels = []


specific_patterns = [
    "Wechat Group",
    "New Boda Safety Check",
    "TS-LPP",
    "TS-RBI",
    "Sales Referral",
    "Whatsapp Group",
    "Waterrate:PM",
    "Waterrate:LL",
    "OCFee: PM",
    "OCFee: LL",
    "MTH",
    "Safetycheck:PM",
    "Safetycheck:LL",
    "MinimumStd: PM",
    "MinimumStd: LL",
    "GR8 Michael Safety Check",
    "myco dnc",
    "Elixir",
    "Channel Agents",
    "Special MF",
    "Partially furnished",
    "Utility 240/M",
    "Furnished",
    "5% Elixir",
    "Elixir 5%"
]


def split_property_labels(label_string):
    
    extracted_items = []
    
    for pattern in specific_patterns:
        if pattern in label_string:
            extracted_items.append(pattern)
            label_string = label_string.replace(pattern, '')  # Remove the matched pattern from the string
    
    
    regex_pattern = (
        r"Safetycheck:PM|Safetycheck:LL|Councilrate:PM|Councilrate:LL|"
        r"MinimumStd: PM|MinimumStd: LL|(?:\d+\.\d+%|Utility \d+/\w+)|"
        r"(?:[A-Za-z'\-\s]+(?:Check|Referral|Safetycheck|Group|Project|account|Property|pending|lease|Michael|Boda|WhatsApp|MTH|Elixir|furnished))|"
        r"(?:\d+\.\d+%)|[^\s]+?:[A-Za-z]+"
    )
    
    
    regex_items = re.findall(regex_pattern, label_string)
    
    
    return extracted_items + regex_items


current_timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')


for index, row in property_labels.iterrows():
    labels = row['Property Labels']
    property_id = row['properties_row']
    split_items = split_property_labels(labels)
    for item in split_items:
        split_labels.append([property_id, item.strip(), current_timestamp, current_timestamp])


split_labels_df = pd.DataFrame(split_labels, columns=['property_id', 'labels', 'created_at', 'updated_at'])


split_labels_df['id'] = range(1000, len(split_labels_df) + 1000)


split_labels_df = split_labels_df[['id', 'property_id', 'labels', 'created_at', 'updated_at']]


output_file_path = r'C:\Users\MSI\Documents\Migration Tool New\DB Tables\Contacts DB\properties_labels.xlsx'
split_labels_df.to_excel(output_file_path, index=False)

print(f"Data has been split and saved to {output_file_path}")
```
